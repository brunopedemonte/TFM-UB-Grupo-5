```python
# Convert variables to numeric
numeric_cols = ['gross_orders', 'cancelled_orders']
result[numeric_cols] = result[numeric_cols].apply(pd.to_numeric)
# Fill NA values with 0 in column 'A'
result['cancelled_orders'].fillna(0, inplace=True)
# Value >0
result = result.loc[result['gross_orders'] >= 0]
result = result.loc[result['cancelled_orders'] >= 0]
# Net Online Orders
result['net_online_orders'] = result['gross_orders'] - result['cancelled_orders']
result.head()
```

```python
df.payment_method_name.unique()
```

```python
# Create the new column 'new_column' based on the condition
result['pay_at_venue'] = result['payment_method_name'].apply(lambda x: 1 if x == 'Pay at venue' else 0)
result.head()
```

```python
# Order Cancellation Rate
result['cancellation_rate'] = result['cancelled_orders'] / result['gross_orders']
result.head(5)
```

```python
# Convert variables to numeric
numeric_cols = ['gross_revenue_eur', 'cancelled_revenue_eur']
result[numeric_cols] = result[numeric_cols].apply(pd.to_numeric)
# Fill NA values with 0 in column 'A'
result['cancelled_revenue_eur'].fillna(0, inplace=True)
# Value >0
result = result.loc[result['gross_revenue_eur'] >= 0]
result = result.loc[result['cancelled_revenue_eur'] >= 0]
# Net Revenue
result['net_revenue'] = result['gross_revenue_eur'] - result['cancelled_revenue_eur']
result.head()
```

```python
# Revenue Cancellation Rate
result['revenue_cancellation_rate'] = result['cancelled_revenue_eur'] / result['gross_revenue_eur']
result['revenue_cancellation_rate'].fillna(0, inplace=True)
result.head()
```

```python
# Convert variables to numeric
numeric_cols = ['gross_item_eur_amount_ttv', 'cancelled_item_eur_amount_ttv']
result[numeric_cols] = result[numeric_cols].apply(pd.to_numeric)
# Fill NA values with 0 in column 'A'
result['cancelled_item_eur_amount_ttv'].fillna(0, inplace=True)
# Value >0
result = result.loc[result['gross_item_eur_amount_ttv'] >= 0]
result = result.loc[result['cancelled_item_eur_amount_ttv'] >= 0]
# Net TTV
result['net_TTV'] = result['gross_item_eur_amount_ttv'] - result['cancelled_item_eur_amount_ttv']
result.head()
```

```python
# Net AOV
result['net_AOV'] = result['net_TTV'] / result['net_online_orders']
result['net_AOV'].fillna(0, inplace=True)
result.head()
```

```python
# Net Take Rate
result['net_Take_Rate'] = result['net_revenue'] / result['net_TTV']
result['net_Take_Rate'].fillna(0, inplace=True)
result.head()
```
