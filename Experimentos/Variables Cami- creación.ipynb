```python

df = pd.read_csv("/content/drive/MyDrive/TFM/Raw_CSV/df_new.csv", sep="	", decimal=".", na_values=['(null)'], low_memory=False)
df.head()
```

```python
((df.isnull().sum() / len(df))*100).sort_values(ascending = False)
```

```python
df.info()
```

```python
df= df.loc[:, ['date_day', 'unique_venue_id', 'country_code','saas_product', 'first_live_date', 'latest_live_date'  ,'new_latest_live_date']]
```

```python
#preparacion de la data: fechas
df['latest_live_date'] = pd.to_datetime(df['latest_live_date'], errors='coerce')
df['new_latest_live_date'] = pd.to_datetime(df['new_latest_live_date'], errors='coerce')
df['first_live_date'] = pd.to_datetime(df['first_live_date'], errors='coerce')
df['date_day'] = pd.to_datetime(df['date_day'])

#trabajo de nulos
# Si latest_live_date es nulo, seleccionamos first_live_date
df['new_latest_live_date'].fillna(df['first_live_date'], inplace=True)
```

```python
df.sample(10)
```

# Variables originales

```python
#mapear country dada la poca frecuencia de los 3 ultimos paises
df.groupby('country_code')['unique_venue_id'].nunique().sort_values(ascending=False)
```

```python
#pd.unique(df['country_code']).tolist()
country_list=  ['GB', 'FR', 'IT', 'AT', 'IE', 'CH', 'DE', 'ES', 'NL',  'BE',  'LT', 'PT', 'GR']

df['country'] = df['country_code'].map(lambda x: 'other' if x not in country_list else x)
df.groupby('country')['unique_venue_id'].nunique().sort_values(ascending=False)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder = OneHotEncoder( sparse_output=False)#dado que se trabaja con modelos de ensamblado, no es necesario tener n-1 variables dicotomicas
onehot_encoded = onehot_encoder.fit_transform(df.loc[:, ['country','saas_product']])
df2= pd.DataFrame(onehot_encoded, columns=onehot_encoder.get_feature_names_out())
df2.sample(10)
```

# Variables nuevas

```python
#creacion de la variable
def calcular_meses(df):
  meses= (df['date_day'].year- df['new_latest_live_date'].year)*12 + (df['date_day'].month- df['new_latest_live_date'].month)
  return meses

df['antiguedad'] = df.apply(calcular_meses, axis=1)
```

```python
def calcular_reactivada(df):
  '''
  Retorna 1 o 0 si la venue es reactivada o nunca se ha ido.
  Para que una venue se considere reactivada debe existir al menos 30 dias de diferencia entre su primer live date(first_live_date) y la version actualizada en la variable 'latest_live_date'

  '''
  dias= abs(df['first_live_date'].day- df['new_latest_live_date'].day)
  if dias>30:
    return 1
  return 0

df['reactivada'] = df.apply(calcular_reactivada, axis=1)
```

```python
df_final= pd.concat([df,df2],axis=1)
df_final.drop(columns=['country_code', 'country', 'saas_product', 'new_latest_live_date','latest_live_date', 'first_live_date'], inplace= True)
df_final.head()
```
