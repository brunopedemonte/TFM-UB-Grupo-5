```python
# Importamos las librer√≠as
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings

# Conectar con Drive
from google.colab import drive
drive.mount('/content/drive')
```

# Cargamos Datasets


```python
dataset_cami_orders = pd.read_csv("/content/drive/Shareddrives/Data_TFM/Dataset_Final/dataset_cami_orders.csv", sep="	", decimal=".")
# Seleccionar las filas donde unique_venue_id no es igual a -1
dataset_cami_orders = dataset_cami_orders[dataset_cami_orders['unique_venue_id'] != -1]
dataset_cami_orders.head(2)
```

```python
dataset_cami = pd.read_csv("/content/drive/Shareddrives/Data_TFM/Dataset_Final/dataset_cami.csv", sep="	", decimal=".")
dataset_cami['date_day'] = pd.to_datetime(dataset_cami['date_day'])
dataset_cami["date_day"] = dataset_cami['date_day'] - dt.timedelta(weeks=4)
dataset_cami['date_day'] = dataset_cami['date_day'].dt.strftime('%Y-%m-01')
dataset_cami['date_day'] = dataset_cami['date_day'].astype('object')
dataset_cami.head(2)
```

```python
df_orders_bruno_uri = pd.read_csv("/content/drive/Shareddrives/Data_TFM/Dataset_Final/df_orders_bruno_uri.csv", sep=",", decimal=".")
df_orders_bruno_uri.head(2)
```

```python
df_venues_bruno = pd.read_csv("/content/drive/Shareddrives/Data_TFM/Dataset_Final/df_venues_bruno.csv", sep=",", decimal=".", low_memory=False)
df_venues_bruno['date_day'] = pd.to_datetime(df_venues_bruno['date_day'])
df_venues_bruno["date_day"] = df_venues_bruno['date_day'] - dt.timedelta(weeks=4)
df_venues_bruno['date_day'] = df_venues_bruno['date_day'].dt.strftime('%Y-%m-01')
df_venues_bruno['date_day'] = df_venues_bruno['date_day'].astype('object')
df_venues_bruno.head(2)
```

```python
orders_arnau = pd.read_csv("/content/drive/Shareddrives/Data_TFM/Dataset_Final/orders_arnau.csv", sep=",", decimal=".")
orders_arnau.head(2)
```

```python
uri_venues = pd.read_csv("/content/drive/Shareddrives/Data_TFM/Dataset_Final/uri_venues.csv", sep="	", decimal=".")
uri_venues['date_day'] = pd.to_datetime(uri_venues['date_day'])
uri_venues["date_day"] = uri_venues['date_day'] - dt.timedelta(weeks=4)
uri_venues['date_day'] = uri_venues['date_day'].dt.strftime('%Y-%m-01')
uri_venues['date_day'] = uri_venues['date_day'].astype('object')
uri_venues.head(2)
```

```python
venue_arnau = pd.read_csv("/content/drive/Shareddrives/Data_TFM/Dataset_Final/venue_arnau.csv", sep=",", decimal=".")
venue_arnau['date_day'] = pd.to_datetime(venue_arnau['date_day'])
venue_arnau["date_day"] = venue_arnau['date_day'] - dt.timedelta(weeks=4)
venue_arnau['date_day'] = venue_arnau['date_day'].dt.strftime('%Y-%m-01')
venue_arnau['date_day'] = venue_arnau['date_day'].astype('object')
venue_arnau.head(2)
```

# Merge Datasets

```python
df_final = pd.merge(dataset_cami, dataset_cami_orders,  how='left', left_on=['unique_venue_id','date_day'], right_on = ['unique_venue_id','date_day'])
df_final = pd.merge(df_final, df_orders_bruno_uri,  how='left', left_on=['unique_venue_id','date_day'], right_on = ['unique_venue_id','date_day'])
df_final = pd.merge(df_final, df_venues_bruno,  how='left', left_on=['unique_venue_id','date_day'], right_on = ['unique_venue_id','date_day'])
df_final = pd.merge(df_final, orders_arnau,  how='left', left_on=['unique_venue_id','date_day'], right_on = ['unique_venue_id','date_day'])
df_final = pd.merge(df_final, uri_venues,  how='left', left_on=['unique_venue_id','date_day'], right_on = ['unique_venue_id','date_day'])
df_final = pd.merge(df_final, venue_arnau,  how='left', left_on=['unique_venue_id','date_day'], right_on = ['unique_venue_id','date_day'])
df_final.head(2)
```

```python
# B√∫squeda de NA
df_final.isna().sum()
```

```python
# Eliminamos filas con Antiguedad y Churn Nulo --> Se eliminan alrededor de 1M de Observaciones
df_final = df_final[df_final['antiguedad'].notna()]
```

```python
# En el caso de las variables de Appointments y Orders y Segmentos, sin es NA, es 0.
df_final.loc[:, "appointments_l30d"] = df_final["appointments_l30d"].fillna(0)
df_final.loc[:, "online_appointments_rate"] = df_final["online_appointments_rate"].fillna(0)
df_final.loc[:, "net_orders"] = df_final["net_orders"].fillna(0)
df_final.loc[:, "orders_cancellation_rate"] = df_final["orders_cancellation_rate"].fillna(0)
df_final.loc[:, "net_aov"] = df_final["net_aov"].fillna(0)
df_final.loc[:, "net_take_rate"] = df_final["net_take_rate"].fillna(0)
df_final.loc[:, "Newbie"] = df_final["Newbie"].fillna(0)
df_final.loc[:, "Referral"] = df_final["Referral"].fillna(0)
df_final.loc[:, "Explorer"] = df_final["Explorer"].fillna(0)
df_final.loc[:, "Loyal"] = df_final["Loyal"].fillna(0)
```

```python
# Si Venue Type Name is nulo, Beauty Salon
df_final.loc[:, "venue_type_name"] = df_final["venue_type_name"].fillna("Beauty Salon")
```

```python
# B√∫squeda de NA
df_final.isna().sum()
```

```python
# Appointments siempre deber ser mayor que orders
(df_final["appointments_l30d"] < df_final["net_orders"]).sum()
```

```python
# En estos casos, imputamos el valor de total orders = Appointments para cumplir con la logica de negocio
mask = df_final["appointments_l30d"] < df_final["net_orders"]
df_final.loc[mask, "appointments_l30d"] = df_final.loc[mask, "net_orders"] / (1 - df_final.loc[mask, "orders_cancellation_rate"])
(df_final["appointments_l30d"] < df_final["net_orders"]).sum()
```

```python
# No pueden tener 2 SaaS funcionando al mismo tiempo
(df_final["saas_product_tw_connect"] == df_final["saas_product_tw_pro"]).sum()
```

```python
# Imputamos prepay con el valor que es la Moda del Dataset
# Reemplazar "NA" por 0 en la columna "prepay"
df_final['prepay'] = df_final['prepay'].fillna(0)
```
